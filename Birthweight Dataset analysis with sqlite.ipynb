```python
def hailstone(x):
    step = 0
    while x != 1:
        step += 1
        x = x//2 if x%2==0 else 3*x+1
    return step
```

```python
con.create_function('hs',1,hailstone)
c.execute('SELECT hs(cigs), cigs FROM btable WHERE cigs>0;')
for item in c.fetchall():
    print(item)
```

```python
class geomean:
    def __init__(self):
        self.sum = 0
        self.count = 0
    def step(self,value):
        self.sum += np.log(value)
        self.count += 1
    def finalize(self):
        return np.exp(self.sum/self.count)
```

```python
class geomean2:
    def __init__(self):
        self.prod = 1.0
        self.count = 0
    def step(self,value):
        self.prod *= value
        self.count += 1
    def finalize(self):
        return self.prod**(1/self.count)
```

```python
con.create_aggregate('geo',1,geomean)
c.execute('SELECT geo(bwght), cigs FROM btable GROUP BY cigs;')
for item in c.fetchall():
    print(item)
```

```python
con.create_aggregate('geo2',1,geomean2)
c.execute('SELECT geo2(bwght), cigs, COUNT(cigs) FROM btable GROUP BY cigs;')
for item in c.fetchall():
    print(item)
```

```python
import sqlite3

con = sqlite3.connect("bwght.db")
con.isolation_level = None
cur = con.cursor()

buffer = ""

print("Enter your SQL commands to execute in sqlite3.")
print("Enter a blank line to exit.")

while True:
    line = input()
    if line == "":
        break
    buffer += line
    if sqlite3.complete_statement(buffer):
        try:
            buffer = buffer.strip()
            cur.execute(buffer)

            if buffer.lstrip().upper().startswith("SELECT"):
                print(cur.fetchall())
        except sqlite3.Error as e:
            print("An error occurred:", e.args[0])
        buffer = ""

con.close()
```
